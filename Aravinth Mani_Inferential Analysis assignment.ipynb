# 8)Test the similarity between the degree_t (Sci&Tech) and specialisation (Mkt&HR) with respect to salary  at significance level of 5%.(Make decision using Hypothesis Testing)

```python


# Assuming 'dataset' is your DataFrame
sci_tech_salary = dataset[dataset['degree_t'] == 'Sci&Tech']['salary']
mkt_hr_salary = dataset[dataset['specialisation'] == 'Mkt&HR']['salary']
```

```python
import scipy.stats as stats
# Perform t-test
t_statistic, p_value = stats.ttest_ind(sci_tech_salary, mkt_hr_salary)

# Set significance level
alpha = 0.05

# Output Results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

```python
# Make Decision
if p_value <= alpha:
    print("Reject the null hypothesis: The average salary for Sci&Tech and Mkt&HR is different.")
else:
    print("Fail to reject the null hypothesis: The average salary for Sci&Tech and Mkt&HR is the same.")
```

## Reject the null hypothesis: The average salary for Sci&Tech and Mkt&HR is different.

# 10)What is the probability Density Function of the salary range from 700000 to 900000?

```python
import seaborn as sns
sns.distplot(dataset["salary"])
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```

## The density of salary between 70K to 90K is 0.0005973310593974868

# 11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

```python
sci_tech_data = dataset[dataset['degree_t'] == 'Sci&Tech'][['etest_p', 'mba_p']]

# Drop rows with missing values
sci_tech_data = sci_tech_data.dropna()

# Reset the index and add a new column 'degree_category'
sci_tech_data.reset_index(drop=True, inplace=True)
sci_tech_data['degree_category'] = 'Sci&Tech'
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'degree_category ~ etest_p * mba_p'
model = ols(formula, data=sci_tech_data).fit()


anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
anova_table
```

## The Probability is greater than 0.05. There is significant difference / reject alternate hypothesis

# Additional Question--> T-Test - explanation with example

T- Test is is used to find the similarity between 2 groups.

It can find the similarity based on pvalue in T-Test.
If P-Value is low then there is more similarity (<0.05)
If p-value is high there less similarity.(>0.05)
Genarally in Indusctrial practice, P-Value has been set as 5% i.e., 0.05
2 Types:
Independent (Unpaired) T-Test - Different group same condition
Dependent (Paired) T-test - same group different condition

While writing syntax for T-test, we need to make sure that group are categorical data and condition is Numerical data
